```python
import pandas as pd
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

## 형변환

```python
train_data = TabularDataset(train.drop(['ID'], axis=1))
test_data = TabularDataset(test.drop(['ID'], axis=1))
```

## 학습

```python
predictor = TabularPredictor(label='전화해지여부', eval_metric='f1_macro',).fit(train_data)
```

## 모델 훈련 성능 평가

```python
ld_board = predictor.leaderboard(train_data, silent=True)
```

```python
ld_board
```

## 예측

```python
pred_y = predictor.predict(test)
```

# 제출 파일

```python
submit = pd.DataFrame()
submit['ID'] = test['ID']
submit['전화해지여부'] = pred_y
submit.to_csv('data/submission_model02.csv', index=False)
```
